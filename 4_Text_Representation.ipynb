```python
import pandas as pd
import re
```

```python
df = pd.read_csv("datasets\IMDB Dataset.csv",nrows=100)
```

```python
df.head()
```

### preprocess the dataset

```python
def preprocess(text):
    #lowercase
    text = text.lower()
    
    # Removing punctuation and numbers and HTML tag
    text = re.sub(r'[^\w\s]','',text)
    text = re.sub(r'\d+','',text)
    text = re.sub(r'<[^>]+>', '', text)
    
    return text
```

```python
df['review'] = df['review'].apply(preprocess)
```

```python
df.head()
```

```python
df.shape
```

### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

# Example for structured categorical data
encoder = OneHotEncoder()
one_hot_matrix = encoder.fit_transform([[1], [2], [3]])  # For categorical numbers or labels
print(one_hot_matrix.toarray()) 
```

```python
# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Fit and transform the 'review' column
encoded_array = encoder.fit_transform(df[['review']])  # Pass as a 2D array

# Create a DataFrame for the one-hot encoded matrix
one_hot_df = pd.DataFrame(encoded_array)

# Combine with the original DataFrame
one_hot_df = pd.concat([df, one_hot_df], axis=1)

# Display the resulting DataFrame
one_hot_df.head()
```

### Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
bag_of_words = vectorizer.fit_transform(df["review"])

# Get the feature names (vocabulary)
vocabulary = vectorizer.get_feature_names_out()
print("Vocabulary:", vocabulary)

# Convert the BoW matrix to a dense array for better readability
bow_array = bag_of_words.toarray()

# Add the BoW representation back to the DataFrame
bow_df = pd.DataFrame(bow_array, columns=vocabulary)
bow_df = pd.concat([df, bow_df], axis=1)
bow_df.head()
```

### n-gram

```python
# Initialize CountVectorizer for trigrams
vectorizer = CountVectorizer(ngram_range=(3, 3))  # 3-grams

# Fit and transform the 'review' column
X = vectorizer.fit_transform(df['review'])

# Get the trigram feature names
trigram_features = vectorizer.get_feature_names_out()

# Convert the sparse matrix to a DataFrame for readability
trigram_df = pd.DataFrame(X.toarray(), columns=trigram_features)

# Display the resulting trigram DataFrame
trigram_df.head()
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# Fit and transform the 'review' column
tfidf_matrix = vectorizer.fit_transform(df['review'])

# Get the feature names (vocabulary)
feature_names = vectorizer.get_feature_names_out()

# Convert the TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Display the resulting TF-IDF DataFrame
tfidf_df.head()
```

### Word2Vec

```python
import nltk
nltk.download('punkt')
```

```python
# Word Tokenization function
def word_tokenize_text(text):
    return nltk.word_tokenize(text)

# Sentence Tokenization function
def sentence_tokenize_text(text):
    return nltk.sent_tokenize(text)
```

```python
df['word_tokens'] = df['review'].apply(word_tokenize_text)
```

```python
df.head()
```

```python
from gensim.models import Word2Vec

# Step 2: Train a Word2Vec model
# Define model parameters
word2vec_model = Word2Vec(
    sentences=df['word_tokens'],  # List of tokenized sentences
    vector_size=100,           # Dimensionality of word embeddings
    window=5,                  # Context window size
    min_count=1,               # Ignore words with a frequency lower than this
    workers=4,                 # Number of parallel workers
    sg=0                       # 0 for CBOW, 1 for Skip-gram
)
```

```python
# Example: Compute the mean vector for each review
df['review_vector'] = df['word_tokens'].apply(
    lambda words: sum(word2vec_model.wv[word] for word in words if word in word2vec_model.wv) / len(words)
)
df['review_vector'][0]
```
